PREVIOUS STUDIES VS PROPOSED SYSTEM


```python
import matplotlib.pyplot as plt

systems = [
    "Previous Studies",
    "Proposed System"
]

accuracy = [
    95,        # literature average
    97.885     # your best model (Random Forest)
]

colors = ["orange", "green"]

plt.figure(figsize=(7,5))

bars = plt.bar(
    systems,
    accuracy,
    color=colors
)

plt.title("Previous vs Proposed System Accuracy")

plt.ylabel("Accuracy (%)")

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height:.2f}%',
        ha='center'
    )

plt.show()
```

Model-wise Accuracy Comparison

```python
models = [
    "Random Forest",
    "XGBoost",
    "Decision Tree",
    "Logistic Regression"
]

accuracy = [
    97.885,
    97.68,
    96.395,
    95.5
]

colors = [
    "green",
    "blue",
    "orange",
    "red"
]

plt.figure(figsize=(10,6))

bars = plt.bar(
    models,
    accuracy,
    color=colors
)

plt.title("Accuracy of Machine Learning Models")

plt.xlabel("Models")

plt.ylabel("Accuracy (%)")

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height,
        f'{height:.2f}%',
        ha='center',
        va='bottom'
    )

plt.show()
```

Model-wise Accuracy Comparison

```python
import numpy as np

features = [
    "Multiple Models",
    "Visualization",
    "Explainability",
    "NASA Data",
    "CNN Support"
]

previous = [1, 1, 0, 0, 0]

proposed = [1, 1, 1, 1, 1]

x = np.arange(len(features))

width = 0.35

plt.figure(figsize=(9,5))

plt.bar(
    x - width/2,
    previous,
    width,
    label="Previous Systems",
    color="gray"
)

plt.bar(
    x + width/2,
    proposed,
    width,
    label="Proposed System",
    color="green"
)

plt.xticks(
    x,
    features,
    rotation=30
)

plt.title("Feature Comparison Between Systems")

plt.legend()

plt.ylabel("Availability")

plt.show()
```
